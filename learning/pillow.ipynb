```python
# Open and load an image
image = Image.open("img.png")
image
```

```python
print('image format:',image.format)
```

```python
print('image mode:',image.mode)
```

```python
print('image size:',image.size)
```

```python
# Convert the image to grayscale
gray_image=image.convert('L')
gray_image
```

```python
# Apply a blur filter
blurred_image = image.filter(ImageFilter.BLUR)
blurred_image
```

```python
# Create a new image with a red rectangle
rectangle_image = image.copy()
draw = ImageDraw.Draw(rectangle_image)
draw.rectangle([(50, 50), (200, 150)], outline="red")
rectangle_image
```

```python
# Add text to the image
text_image = image.copy()
draw = ImageDraw.Draw(text_image)
font = ImageFont.truetype("arial.ttf", 20)
draw.text((50, 50), "Hello, PIL!", fill="black", font=font)
text_image
```

```python
# Resize the image
image = image.resize((400, 300))
image
```

```python
# Save the modified images
gray_image.save("output_gray.jpg")
blurred_image.save("output_blurred.jpg")
rectangle_image.save("output_rectangle.jpg")
text_image.save("output_text.jpg")
```
