```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.wikipedia.org'
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
title=soup.title
title.text
```

```python
print(soup.prettify())
```

```python
paras=soup.find_all('p',limit=1)
for para in paras:
    print(para.text)
```

```python
links=soup.find_all('a')
for link in links:
    print(link['href'])
```

```python
li=soup.li
next_sibling_li=li.find_next_sibling()
next_sibling_li
```

```python
body=soup.body
previous_sibling_body=body.find_previous_sibling()
previous_sibling_body
```

```python
ul=soup.ul
child_ul=ul.children
for child in child_ul:
    print(child.text)
```

```python
div=soup.find('div',class_="central-featured-lang lang8")
div
```
